###  Q16. Which sub category had highest growth by profit in 2023 compare to 2022

```python
m = df1.groupby(['Sub Category', 'year']).sum(['Total Profit'])['Total Profit'].reset_index()
m
```

```python
n = m.pivot_table(index = 'Sub Category', columns='year', values = 'Total Profit').reset_index()
n
```

```python
n['diff'] = n[2023] - n[2022]
```

```python
n.sort_values(by = ['diff'], ascending=False).head(1)
```
